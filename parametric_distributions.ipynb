```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Simulate a random sample from a Bernoulli distribution with p=0.5
p = 0.5
sample = stats.bernoulli.rvs(p, size=100)

plt.title(f"Histogram of a sample from Bernoulli({p})")
plt.hist(sample, bins=20)
plt.legend()
plt.show()
plt.close()
```

```python
# Get sample mean of Bernoulli sample and plot histogram of sample with vertical line for sample mean 
sample_mean = np.mean(sample)
plt.title(f"Histogram of a sample from Bernoulli({p}) with vertical line for sample mean")
plt.hist(sample, bins=20)
plt.axvline(sample_mean, color='red', linestyle='--', label='Sample Mean')
plt.legend()
plt.show()
plt.close()
```

```python
# Simulate a random sample from a normal distribution with mean 0 and standard deviation 1
sample = stats.norm.rvs(0, 1, size=100)

plt.title("Histogram of a sample from N(0,1)")
plt.hist(sample, bins=30, density=True)
plt.show()
plt.close()

# Plot the histogram of the sample
plt.title("Histogram of a sample from N(0,1)")
plt.hist(sample, bins=30, density=True, alpha=0.5, color='blue', label='Sample')
# Plot the normal distribution
x = np.linspace(-4, 4, 1000)
plt.plot(x, stats.norm.pdf(x, 0, 1), color='red', label='Normal Distribution') 
plt.legend()
plt.show()
plt.close()
```

```python
# Estimating the mean and standard deviation of the sample
mean = np.mean(sample)
std = np.std(sample)
print(f"Mean: {mean}, Standard Deviation: {std}")

# Plotting the sample mean and standard deviation
plt.figure(figsize=(10, 6))
plt.title("Sample Mean and Standard Deviation")
plt.plot(x, stats.norm.pdf(x, 0, 1), color='blue', label='True Normal Distribution')
plt.plot(x, stats.norm.pdf(x, mean, std), color='red', label='Estimated Normal Distribution') 
plt.axvline(mean, color='blue', linestyle='--', label='Sample Mean')
plt.axvline(mean + std, color='green', linestyle='--', label='Sample Mean + STDev')
plt.axvline(mean - std, color='green', linestyle='--', label='Sample Mean - STDev')
plt.legend()
plt.show()
plt.close()
```
